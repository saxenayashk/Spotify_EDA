```python
#importing the libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Reading and displaying the csv file
data=pd.read_csv("C:\\Users\\Yash Saxena\\Desktop\\Spotify_Data_Analysis\\tracks.csv")
data.head()
```

```python
#checking null values
data.isnull().sum()
```

```python
#Checking the number of rows and columns
data.info()
```

```python
#10 Least popular songs on Spotify
sorted_data=data.sort_values('popularity',ascending=True).head(10)
print(sorted_data)
```

```python
data.describe().transpose()
```

```python
#Most popular songs on Spotify
most_popular=data.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
#Converting the duration from milliseconds to seconds
data["duration"]=data["duration_ms"].apply(lambda x:round(x/1000))
data.drop("duration_ms",inplace=True,axis=1)
```

```python
data.duration.head(10)
```

```python
#Correlation amongst different columns in the dataset
corr_data=data.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_data,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#Finding number of rows in 1% of the dataset
sample_data=data.sample(int(0.01*len(data)))
print(len(sample_data))
```

```python
#Plotting Regression graph between popularity and acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y="popularity",x="acousticness",color="b").set(title="Popularity Vs Acousticness Correlation")
```

```python
#Plotting Regression graph between popularity and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_data,y="loudness",x="energy",color="c").set(title="Loudness Vs Energy Correlation")
```

```python
#Reading the genre csv file
data_genre=pd.read_csv("C:\\Users\\Yash Saxena\Desktop\Spotify_Data_Analysis\SpotifyFeatures.csv")
data_genre.head(10)
```

```python
plt.title("Duration of the songs in Different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=data_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=data_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set("Top 5 Genres by Popularity")
```
